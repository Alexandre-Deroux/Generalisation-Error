# **Les méthodes de généralisation**

## Régression linéaire :

Régression linéaire (Linear Regression) : La régression linéaire est utilisée pour modéliser la relation linéaire entre une variable cible continue et un ensemble de variables explicatives. Le modèle apprend les coefficients de régression qui minimisent l'erreur quadratique moyenne entre les prédictions et les valeurs réelles.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('german_credit_risk_21.csv', sep=',')

# Liste des colonnes contenant du texte
text_columns = df.select_dtypes(include=['object']).columns

# Remplacement du texte par des valeurs numériques
for column in text_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))

print(df.head())

X = df.drop('Credit Amount', axis=1)

Y = df.loc[:, 'Credit Amount']

# Charger les données d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialiser le modèle
model = LinearRegression()

scaler_Norm = MinMaxScaler()
x_train_Norm = scaler_Norm.fit_transform(X_train)
x_test_Norm = scaler_Norm.fit_transform(X_test)

# Entraîner le modèle sur les données d'entraînement
model.fit(x_train_Norm, y_train)

# Faire des prédictions sur les données de test
y_pred = model.predict(x_test_Norm)

# Calculer l'erreur de généralisation (RMSE)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R-squared:", r2)
```

## Réseaux de neurones :



Réseaux de neurones (Neural Networks) : Les réseaux de neurones sont des modèles d'apprentissage profond qui simulent le fonctionnement du cerveau humain. Ils sont composés de couches de neurones interconnectés et sont capables d'apprendre des représentations complexes des données. Les réseaux de neurones sont utilisés pour diverses tâches, telles que la classification, la détection d'objets, la traduction automatique, etc.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('german_credit_risk_21.csv', sep=',')

# Liste des colonnes contenant du texte
text_columns = df.select_dtypes(include=['object']).columns

# Remplacement du texte par des valeurs numériques
for column in text_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))

print(df.head())

X = df.drop('Credit Amount', axis=1)

Y = df.loc[:, 'Credit Amount']

# Charger les données d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialiser le modèle de réseau de neurones
model = MLPRegressor(hidden_layer_sizes=(100, 100), activation='relu', solver='adam', random_state=42)

scaler_Norm = MinMaxScaler()
x_train_Norm = scaler_Norm.fit_transform(X_train)
x_test_Norm = scaler_Norm.fit_transform(X_test)

# Entraîner le modèle sur les données d'entraînement
model.fit(x_train_Norm, y_train)

# Faire des prédictions sur les données de test
y_pred = model.predict(x_test_Norm)

# Calculer l'erreur de généralisation (RMSE)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R-squared:", r2)
```

## Machines à vecteurs de support :



Machines à vecteurs de support (Support Vector Machines - SVM) : Les SVM sont des modèles d'apprentissage supervisé utilisés pour la classification et la régression. Les SVM peuvent également utiliser des noyaux pour traiter des problèmes non linéaires.

## Méthodes d'ensemble :



Méthodes d'ensemble (Ensemble Methods) : Les méthodes d'ensemble combinent plusieurs modèles pour améliorer les performances de généralisation. Des exemples de méthodes d'ensemble comprennent les forêts aléatoires (Random Forests), les méthodes de boosting (AdaBoost, Gradient Boosting), les méthodes de bagging (Bootstrap Aggregating), etc.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('german_credit_risk_21.csv', sep=',')

# Liste des colonnes contenant du texte
text_columns = df.select_dtypes(include=['object']).columns

# Remplacement du texte par des valeurs numériques
for column in text_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))

print(df.head())

X = df.drop('Credit Amount', axis=1)

Y = df.loc[:, 'Credit Amount']

# Charger les données d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialiser le modèle de Méthodes d'ensemble (AdaBoost)
model = AdaBoostRegressor(n_estimators=100, random_state=42)

scaler_Norm = MinMaxScaler()
x_train_Norm = scaler_Norm.fit_transform(X_train)
x_test_Norm = scaler_Norm.fit_transform(X_test)

# Entraîner le modèle sur les données d'entraînement
model.fit(x_train_Norm, y_train)

# Faire des prédictions sur les données de test
y_pred = model.predict(x_test_Norm)

# Calculer l'erreur de généralisation (RMSE)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R-squared:", r2)
```

# **Meilleure méthode sans la dimension effective**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('german_credit_risk_21.csv', sep=',')

# Liste des colonnes contenant du texte
text_columns = df.select_dtypes(include=['object']).columns

# Remplacement du texte par des valeurs numériques
for column in text_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))

print(df.head())

X = df.drop('Age (years)', axis=1)

Y = df.loc[:, 'Age (years)']

# Charger les données d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

scaler_STD = StandardScaler()
# scaler_Norm = MinMaxScaler()
X_train = scaler_STD.fit_transform(X_train)
X_test = scaler_STD.fit_transform(X_test)

# Initialiser les modèles
linear_regression = LinearRegression()
logistic_regression = LogisticRegression()
decision_tree = DecisionTreeRegressor()
neural_network = MLPRegressor(hidden_layer_sizes=(100, 100), activation='relu', solver='adam', random_state=42)
support_vector_machine = SVR()
ensemble_methods = AdaBoostRegressor(n_estimators=100, random_state=42)

# Entraîner les modèles
linear_regression.fit(X_train, y_train)
logistic_regression.fit(X_train, y_train)
decision_tree.fit(X_train, y_train)
neural_network.fit(X_train, y_train)
support_vector_machine.fit(X_train, y_train)
ensemble_methods.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred_linear_regression = linear_regression.predict(X_test)
y_pred_logistic_regression = logistic_regression.predict(X_test)
y_pred_decision_tree = decision_tree.predict(X_test)
y_pred_neural_network = neural_network.predict(X_test)
y_pred_support_vector_machine = support_vector_machine.predict(X_test)
y_pred_ensemble_methods = ensemble_methods.predict(X_test)

# Calculer les métriques d'évaluation
rmse_linear_regression = mean_squared_error(y_test, y_pred_linear_regression, squared=False)
rmse_logistic_regression = mean_squared_error(y_test, y_pred_logistic_regression, squared=False)
rmse_decision_tree = mean_squared_error(y_test, y_pred_decision_tree, squared=False)
rmse_neural_network = mean_squared_error(y_test, y_pred_neural_network, squared=False)
rmse_support_vector_machine = mean_squared_error(y_test, y_pred_support_vector_machine, squared=False)
rmse_ensemble_methods = mean_squared_error(y_test, y_pred_ensemble_methods, squared=False)

r2_linear_regression = r2_score(y_test, y_pred_linear_regression)
r2_logistic_regression = r2_score(y_test, y_pred_logistic_regression)
r2_decision_tree = r2_score(y_test, y_pred_decision_tree)
r2_neural_network = r2_score(y_test, y_pred_neural_network)
r2_support_vector_machine = r2_score(y_test, y_pred_support_vector_machine)
r2_ensemble_methods = r2_score(y_test, y_pred_ensemble_methods)

mae_linear_regression = mean_absolute_error(y_test, y_pred_linear_regression)
mae_logistic_regression = mean_absolute_error(y_test, y_pred_logistic_regression)
mae_decision_tree = mean_absolute_error(y_test, y_pred_decision_tree)
mae_neural_network = mean_absolute_error(y_test, y_pred_neural_network)
mae_support_vector_machine = mean_absolute_error(y_test, y_pred_support_vector_machine)
mae_ensemble_methods = mean_absolute_error(y_test, y_pred_ensemble_methods)

mre_linear_regression = np.mean(np.abs((y_test - y_pred_linear_regression) / y_test)) * 100
mre_logistic_regression = np.mean(np.abs((y_test - y_pred_logistic_regression) / y_test)) * 100
mre_decision_tree = np.mean(np.abs((y_test - y_pred_decision_tree) / y_test)) * 100
mre_neural_network = np.mean(np.abs((y_test - y_pred_neural_network) / y_test)) * 100
mre_support_vector_machine = np.mean(np.abs((y_test - y_pred_support_vector_machine) / y_test)) * 100
mre_ensemble_methods = np.mean(np.abs((y_test - y_pred_ensemble_methods) / y_test)) * 100

mean_mse_cv_scores_linear_regression = -(cross_val_score(linear_regression, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_logistic_regression = -(cross_val_score(logistic_regression, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_decision_tree = -(cross_val_score(decision_tree, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_neural_network = -(cross_val_score(neural_network, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_support_vector_machine = -(cross_val_score(support_vector_machine, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_ensemble_methods = -(cross_val_score(ensemble_methods, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()

# Afficher les résultats
print("Performance des modèles :")
print("---------------------------------------------------------------------------------------------------------------------")
print("Méthode \t\t\t RMSE \t\t R-squared \t MAE \t\t MRE (%) \t Mean MSE CV Scores")
print("---------------------------------------------------------------------------------------------------------------------")
print("Régression linéaire \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_linear_regression, r2_linear_regression, mae_linear_regression, mre_linear_regression, mean_mse_cv_scores_linear_regression))
print("Régression logistique \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_logistic_regression, r2_logistic_regression, mae_logistic_regression, mre_logistic_regression, mean_mse_cv_scores_logistic_regression))
print("Arbres de décision \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_decision_tree, r2_decision_tree, mae_decision_tree, mre_decision_tree, mean_mse_cv_scores_decision_tree))
print("Réseaux de neurones \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_neural_network, r2_neural_network, mae_neural_network, mre_neural_network, mean_mse_cv_scores_neural_network))
print("Machines à vecteurs de support \t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_support_vector_machine, r2_support_vector_machine, mae_support_vector_machine, mre_support_vector_machine, mean_mse_cv_scores_support_vector_machine))
print("Méthodes d'ensemble \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_ensemble_methods, r2_ensemble_methods, mae_ensemble_methods, mre_ensemble_methods, mean_mse_cv_scores_ensemble_methods))
print("---------------------------------------------------------------------------------------------------------------------")
```

```python
# Métriques d'évaluation pour chaque méthode
methods = ['Linear Regression', 'Logistic Regression', 'Decision Trees', 'Neural Networks', 'Support Vector Machines', 'Ensemble Methods']
metrics = ['RMSE', 'R2', 'MAE', 'MRE', 'Mean MSE CV Scores']

# Scores de chaque méthode pour chaque métrique
scores = [
    [rmse_linear_regression, r2_linear_regression, mae_linear_regression, mre_linear_regression, mean_mse_cv_scores_linear_regression],
    [rmse_logistic_regression, r2_logistic_regression, mae_logistic_regression, mre_logistic_regression, mean_mse_cv_scores_logistic_regression],
    [rmse_decision_tree, r2_decision_tree, mae_decision_tree, mre_decision_tree, mean_mse_cv_scores_decision_tree],
    [rmse_neural_network, r2_neural_network, mae_neural_network, mre_neural_network, mean_mse_cv_scores_neural_network],
    [rmse_support_vector_machine, r2_support_vector_machine, mae_support_vector_machine, mre_support_vector_machine, mean_mse_cv_scores_support_vector_machine],
    [rmse_ensemble_methods, r2_ensemble_methods, mae_ensemble_methods, mre_ensemble_methods, mean_mse_cv_scores_ensemble_methods]
]

# Spécification pour chaque métrique (minimiser ou maximiser)
metric_specifications = {
    'RMSE': 'min',
    'R2': 'max',
    'MAE': 'min',
    'MRE': 'min',
    'Mean MSE CV Scores': 'min'
}

# Trouver la meilleure méthode pour chaque métrique
best_methods = []
for j in range(len(metrics)):
    metric_scores = [scores[i][j] for i in range(len(methods))]
    if metric_specifications[metrics[j]] == 'min':
        best_index = metric_scores.index(min(metric_scores))
    else:
        best_index = metric_scores.index(max(metric_scores))
    best_method = methods[best_index]
    best_methods.append(best_method)

# Afficher les meilleures méthodes pour chaque métrique
for i in range(len(metrics)):
    print(f"La meilleure méthode pour la métrique '{metrics[i]}' (à {metric_specifications[metrics[i]]}) est : {best_methods[i]}")
```

```python
import matplotlib.pyplot as plt

for i in range(len(metrics)):
    x = range(len(methods))
    scores_metric = [score[i] for score in scores]

    # Trouver les indexes des scores maximum et minimum
    max_index = np.argmax(scores_metric)
    min_index = np.argmin(scores_metric)

    # Création du graphique avec une taille réduite
    plt.figure(figsize=(10, 5))

    # Utilisation de la liste de couleurs pour les colonnes
    colors = ['b'] * len(methods)
    colors[min_index] = 'g'  # Couleur verte pour le score minimum

    if i == 1:  # Pour la métrique R2
        colors[max_index] = 'g'  # Couleur verte pour le score maximum
        colors[min_index] = 'b'  # Couleur bleue pour le score minimum

    plt.bar(x, scores_metric, color=colors)
    plt.xlabel('Methods')
    plt.ylabel('Scores')
    plt.title(f'Scores for the metric "{metrics[i]}"')
    plt.xticks(x, methods, rotation=45)
    plt.tight_layout()

    # Affichage du graphique
    plt.show()
```

# **Meilleure méthode avec ACP**

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('german_credit_risk_21.csv', sep=',')

# Liste des colonnes contenant du texte
text_columns = df.select_dtypes(include=['object']).columns

# Remplacement du texte par des valeurs numériques
for column in text_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))

print(df.head())

X = df.drop('Age (years)', axis=1)

Y = df.loc[:, 'Age (years)']

# Spécifier le nombre de dimensions souhaité
n_dimensions = 6

# Créer une instance de PCA avec le nombre de dimensions spécifié
pca = PCA(n_components=n_dimensions)

# Appliquer la PCA sur les données
X_pca = pca.fit_transform(X)

# Afficher les nouvelles dimensions réduites
print(X_pca)

# Charger les données d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_pca, Y, test_size=0.2, random_state=42)

scaler_STD = StandardScaler()
# scaler_Norm = MinMaxScaler()
X_train = scaler_STD.fit_transform(X_train)
X_test = scaler_STD.fit_transform(X_test)

# Initialiser les modèles
linear_regression = LinearRegression()
logistic_regression = LogisticRegression()
decision_tree = DecisionTreeRegressor()
neural_network = MLPRegressor(hidden_layer_sizes=(100, 100), activation='relu', solver='adam', random_state=42)
support_vector_machine = SVR()
ensemble_methods = AdaBoostRegressor(n_estimators=100, random_state=42)

# Entraîner les modèles
linear_regression.fit(X_train, y_train)
logistic_regression.fit(X_train, y_train)
decision_tree.fit(X_train, y_train)
neural_network.fit(X_train, y_train)
support_vector_machine.fit(X_train, y_train)
ensemble_methods.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred_linear_regression = linear_regression.predict(X_test)
y_pred_logistic_regression = logistic_regression.predict(X_test)
y_pred_decision_tree = decision_tree.predict(X_test)
y_pred_neural_network = neural_network.predict(X_test)
y_pred_support_vector_machine = support_vector_machine.predict(X_test)
y_pred_ensemble_methods = ensemble_methods.predict(X_test)

# Calculer les métriques d'évaluation
rmse_linear_regression = mean_squared_error(y_test, y_pred_linear_regression, squared=False)
rmse_logistic_regression = mean_squared_error(y_test, y_pred_logistic_regression, squared=False)
rmse_decision_tree = mean_squared_error(y_test, y_pred_decision_tree, squared=False)
rmse_neural_network = mean_squared_error(y_test, y_pred_neural_network, squared=False)
rmse_support_vector_machine = mean_squared_error(y_test, y_pred_support_vector_machine, squared=False)
rmse_ensemble_methods = mean_squared_error(y_test, y_pred_ensemble_methods, squared=False)

r2_linear_regression = r2_score(y_test, y_pred_linear_regression)
r2_logistic_regression = r2_score(y_test, y_pred_logistic_regression)
r2_decision_tree = r2_score(y_test, y_pred_decision_tree)
r2_neural_network = r2_score(y_test, y_pred_neural_network)
r2_support_vector_machine = r2_score(y_test, y_pred_support_vector_machine)
r2_ensemble_methods = r2_score(y_test, y_pred_ensemble_methods)

mae_linear_regression = mean_absolute_error(y_test, y_pred_linear_regression)
mae_logistic_regression = mean_absolute_error(y_test, y_pred_logistic_regression)
mae_decision_tree = mean_absolute_error(y_test, y_pred_decision_tree)
mae_neural_network = mean_absolute_error(y_test, y_pred_neural_network)
mae_support_vector_machine = mean_absolute_error(y_test, y_pred_support_vector_machine)
mae_ensemble_methods = mean_absolute_error(y_test, y_pred_ensemble_methods)

mre_linear_regression = np.mean(np.abs((y_test - y_pred_linear_regression) / y_test)) * 100
mre_logistic_regression = np.mean(np.abs((y_test - y_pred_logistic_regression) / y_test)) * 100
mre_decision_tree = np.mean(np.abs((y_test - y_pred_decision_tree) / y_test)) * 100
mre_neural_network = np.mean(np.abs((y_test - y_pred_neural_network) / y_test)) * 100
mre_support_vector_machine = np.mean(np.abs((y_test - y_pred_support_vector_machine) / y_test)) * 100
mre_ensemble_methods = np.mean(np.abs((y_test - y_pred_ensemble_methods) / y_test)) * 100

mean_mse_cv_scores_linear_regression = -(cross_val_score(linear_regression, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_logistic_regression = -(cross_val_score(logistic_regression, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_decision_tree = -(cross_val_score(decision_tree, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_neural_network = -(cross_val_score(neural_network, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_support_vector_machine = -(cross_val_score(support_vector_machine, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()
mean_mse_cv_scores_ensemble_methods = -(cross_val_score(ensemble_methods, X, Y, cv=5, scoring='neg_mean_squared_error')).mean()

# Afficher les résultats
print("Performance des modèles :")
print("---------------------------------------------------------------------------------------------------------------------")
print("Méthode \t\t\t RMSE \t\t R-squared \t MAE \t\t MRE (%) \t Mean MSE CV Scores")
print("---------------------------------------------------------------------------------------------------------------------")
print("Régression linéaire \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_linear_regression, r2_linear_regression, mae_linear_regression, mre_linear_regression, mean_mse_cv_scores_linear_regression))
print("Régression logistique \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_logistic_regression, r2_logistic_regression, mae_logistic_regression, mre_logistic_regression, mean_mse_cv_scores_logistic_regression))
print("Arbres de décision \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_decision_tree, r2_decision_tree, mae_decision_tree, mre_decision_tree, mean_mse_cv_scores_decision_tree))
print("Réseaux de neurones \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_neural_network, r2_neural_network, mae_neural_network, mre_neural_network, mean_mse_cv_scores_neural_network))
print("Machines à vecteurs de support \t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_support_vector_machine, r2_support_vector_machine, mae_support_vector_machine, mre_support_vector_machine, mean_mse_cv_scores_support_vector_machine))
print("Méthodes d'ensemble \t\t {:.4f} \t {:.4f} \t {:.4f} \t {:.4f} \t\t {:.4f}".format(rmse_ensemble_methods, r2_ensemble_methods, mae_ensemble_methods, mre_ensemble_methods, mean_mse_cv_scores_ensemble_methods))
print("---------------------------------------------------------------------------------------------------------------------")
```

```python
# Métriques d'évaluation pour chaque méthode
methods = ['Linear Regression', 'Logistic Regression', 'Decision Trees', 'Neural Networks', 'Support Vector Machines', 'Ensemble Methods']
metrics = ['RMSE', 'R2', 'MAE', 'MRE', 'Mean MSE CV Scores']

# Scores de chaque méthode pour chaque métrique
scores = [
    [rmse_linear_regression, r2_linear_regression, mae_linear_regression, mre_linear_regression, mean_mse_cv_scores_linear_regression],
    [rmse_logistic_regression, r2_logistic_regression, mae_logistic_regression, mre_logistic_regression, mean_mse_cv_scores_logistic_regression],
    [rmse_decision_tree, r2_decision_tree, mae_decision_tree, mre_decision_tree, mean_mse_cv_scores_decision_tree],
    [rmse_neural_network, r2_neural_network, mae_neural_network, mre_neural_network, mean_mse_cv_scores_neural_network],
    [rmse_support_vector_machine, r2_support_vector_machine, mae_support_vector_machine, mre_support_vector_machine, mean_mse_cv_scores_support_vector_machine],
    [rmse_ensemble_methods, r2_ensemble_methods, mae_ensemble_methods, mre_ensemble_methods, mean_mse_cv_scores_ensemble_methods]
]

# Spécification pour chaque métrique (minimiser ou maximiser)
metric_specifications = {
    'RMSE': 'min',
    'R2': 'max',
    'MAE': 'min',
    'MRE': 'min',
    'Mean MSE CV Scores': 'min'
}

# Trouver la meilleure méthode pour chaque métrique
best_methods = []
for j in range(len(metrics)):
    metric_scores = [scores[i][j] for i in range(len(methods))]
    if metric_specifications[metrics[j]] == 'min':
        best_index = metric_scores.index(min(metric_scores))
    else:
        best_index = metric_scores.index(max(metric_scores))
    best_method = methods[best_index]
    best_methods.append(best_method)

# Afficher les meilleures méthodes pour chaque métrique
for i in range(len(metrics)):
    print(f"La meilleure méthode pour la métrique '{metrics[i]}' (à {metric_specifications[metrics[i]]}) est : {best_methods[i]}")
```

```python
import matplotlib.pyplot as plt

for i in range(len(metrics)):
    x = range(len(methods))
    scores_metric = [score[i] for score in scores]

    # Trouver les indexes des scores maximum et minimum
    max_index = np.argmax(scores_metric)
    min_index = np.argmin(scores_metric)

    # Création du graphique avec une taille réduite
    plt.figure(figsize=(10, 5))

    # Utilisation de la liste de couleurs pour les colonnes
    colors = ['b'] * len(methods)
    colors[min_index] = 'g'  # Couleur verte pour le score minimum

    if i == 1:  # Pour la métrique R2
        colors[max_index] = 'g'  # Couleur verte pour le score maximum
        colors[min_index] = 'b'  # Couleur bleue pour le score minimum

    plt.bar(x, scores_metric, color=colors)
    plt.xlabel('Methods')
    plt.ylabel('Scores')
    plt.title(f'Scores for the metric "{metrics[i]}"')
    plt.xticks(x, methods, rotation=45)
    plt.tight_layout()

    # Affichage du graphique
    plt.show()
```

```python
# Métriques d'évaluation pour chaque méthode
methods = ['Linear Regression', 'Logistic Regression', 'Decision Trees', 'Neural Networks', 'Support Vector Machines', 'Ensemble Methods']
metrics = ['RMSE', 'R2', 'MAE', 'MRE', 'Mean MSE CV Scores', 'Generalisation Error']

# Scores de chaque méthode pour chaque métrique
scores = [
    [rmse_linear_regression, r2_linear_regression, mae_linear_regression, mre_linear_regression, mean_mse_cv_scores_linear_regression, generalisation_error_linear_regression],
    [rmse_logistic_regression, r2_logistic_regression, mae_logistic_regression, mre_logistic_regression, mean_mse_cv_scores_logistic_regression, generalisation_error_logistic_regression],
    [rmse_decision_tree, r2_decision_tree, mae_decision_tree, mre_decision_tree, mean_mse_cv_scores_decision_tree, generalisation_error_decision_tree],
    [rmse_neural_network, r2_neural_network, mae_neural_network, mre_neural_network, mean_mse_cv_scores_neural_network, generalisation_error_neural_network],
    [rmse_support_vector_machine, r2_support_vector_machine, mae_support_vector_machine, mre_support_vector_machine, mean_mse_cv_scores_support_vector_machine, generalisation_error_support_vector_machine],
    [rmse_ensemble_methods, r2_ensemble_methods, mae_ensemble_methods, mre_ensemble_methods, mean_mse_cv_scores_ensemble_methods, generalisation_error_ensemble_methods]
]

# Spécification pour chaque métrique (minimiser ou maximiser)
metric_specifications = {
    'RMSE': 'min',
    'R2': 'max',
    'MAE': 'min',
    'MRE': 'min',
    'Mean MSE CV Scores': 'min',
    'Generalisation Error': 'min'
}

# Trouver la meilleure méthode pour chaque métrique
best_methods = []
for j in range(len(metrics)):
    metric_scores = [scores[i][j] for i in range(len(methods))]
    if metric_specifications[metrics[j]] == 'min':
        best_index = metric_scores.index(min(metric_scores))
    else:
        best_index = metric_scores.index(max(metric_scores))
    best_method = methods[best_index]
    best_methods.append(best_method)

# Afficher les meilleures méthodes pour chaque métrique
for i in range(len(metrics)):
    print(f"La meilleure méthode pour la métrique '{metrics[i]}' (à {metric_specifications[metrics[i]]}) est : {best_methods[i]}")
```
